```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# df = pd.read_csv("housing.csv")
```

```python
ds = pd.read_csv("alloy_data.csv")
ds.head()
```

```python
df = pd.DataFrame({
    "Size (sqft)": [1200, 1500, 800, 2000, 5000],
    "Bedrooms": [3, 4, 2, 5, 6],
    "Age (years)": [10, 5, 20, 2, 1],
    "Distance to City (km)": [5, 3, 15, 2, 1],
    "Price (£)": [300000, 450000, 200000, 600000, 2000000]
})
df
```

```python
# Histogram - distribution of a single variable
# How to read it:

# Symmetric → normal distribution
# Skewed right → many small values, few large ones
# Long tail → possible outliers

# 👉 If you see one bar far away from others → likely an outlier
df["Price (£)"].hist()
```

```python
# Boxplot
# A boxplot shows:

# Median (middle line)
# Quartiles (box)
# Whiskers (range)
# Points outside → outliers
sns.boxplot(x=df["Price (£)"])
```

```python
# Scatter plot

# How to read:
# Upward trend → positive correlation
# Downward → negative correlation
# Random → no relationship
# 👉 If one point is far away from trend → outlier
sns.scatterplot(x="Size (sqft)", y="Price (£)", data=df)
```

```python
# Correlation

# Interpretation:

# Close to 1 or -1 → strong relationship
# Close to 0 → weak
sns.heatmap(df.corr(), annot=True)
df.corr()
```

```python
# Pairplot
# Shows scatter plots between all features
# Helps quickly detect:
# correlations
# clusters
# outliers
sns.pairplot(df)
```
